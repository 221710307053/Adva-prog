```
#include<stdio.h>
int main()
{
	int i,j;
	for(i=1;i<=4;i++)//no of rows
	{
		for(j=1;j<=i;j++)//no of chars
		printf("*");
		printf("\n");
	}
	
}

*
**
***
****



#include<stdio.h>
int main()
{
	int i,j;
	for(i=1;i<=4;i++)//no of rows
	{
		for(j=1;j<=2*i-1;j++)//no of chars
		printf("*");
		printf("\n");
	}
	
}

*
***
*****
*******

#include<stdio.h>
int main()
{
	int i,j;
	for(i=1;i<=4;i++)//no of rows
	{
		for(j=1;j<=2*i-1;j++)//no of chars
		printf("*");
		printf("\n");
	}
	for(i=3;i>=1;i++)//no of rows
	{
		for(j=1;j<=2*i-1;j++)//no of chars
		printf("*");
		printf("\n");
	}
}

*
***
*****
*******
*****
***
*

#include<stdio.h>
int main()
{
	int i,j,sp;
	for(i=1;i<=4;i++)//no of rows
	{	
		for(sp=1;sp<=4-i;sp++)
		printf(" ");
		for(j=1;j<=2*i-1;j++)//no of chars
		printf("*");
		printf("\n");
	}
	for(i=3;i>=1;i++)//no of rows
	{
		for(sp=1;sp<=4-i;sp++)
		printf(" ");
		for(j=1;j<=2*i-1;j++)//no of chars
		printf("*");
		printf("\n");
	}
}


                  *
                ***	
              *****
            *******
              *****
                ***
                  *

#include<stdio.h>
int main()
{
	int i,j;
	for(i=1;i<=4;i++)//no of rows
	{
		for(j=1;j<=i;j++)//no of chars
		printf("%d",i);
		printf("\n");
	}
	
}


1
22
333
4444



#include<stdio.h>
int main()
{
	int i,j;
	for(i=1;i<=4;i++)//no of rows
	{
		for(j=1;j<=i;j++)//no of chars
		printf("%d",j);
		printf("\n");
	}
	
}



1
12
123
1234


#include<stdio.h>
int main()
{
	int i,j,count=1;
	for(i=1;i<=4;i++)//no of rows
	{
		for(j=1;j<=i;j++)//no of chars
		printf("%d",count++);
		printf("\n");
	}
	
}

1
23
456
78910

#include<stdio.h>
int main()
{
	int i,j,k,l;
	//printing above triangle
	for(i=1;i<=4;i++)//no of rows
	{
		for(k=1;k<=4-i;k++)//space loop
		printf(" ");
		for(j=1;j<=i;j++)//no of chars on left triangle
		printf("%d",j);
		for(l=i-1;l>=1;l--)//no of chars on right triangle
		printf("%d",l);
		printf("\n");
	}
	//printing downward triangle
	for(i=3;i>=1;i--)
	{
		for(k=1;k<=4-i;k++)
		printf(" ");
		for(j=1;j<=i;j++)
		printf("%d",j);
		for(l=i-1;l>=1;l--)
		printf("%d",l);
		printf("\n");
	}
}



                  1
                121
              12321
            1234321
              12321
                121
                  1
```
